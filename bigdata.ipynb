```python
 import matplotlib.pyplot as xyz
 import requests as rq
 import pandas as pd
 import numpy as np
 plotdata = pd.DataFrame({
 "Jan" :[20, 35, 30, 35, 27],
 "Feb" : [25, 40, 27, 10, 15],
 "Mar" : [15, 32, 30, 30, 10],
 "Apr" : [30, 12, 34, 20, 25]},
 index = ["A","B","C","D","E"])
 plotdata.plot(kind='bar', stacked=True, width=0.35)
 xyz.ylabel('Number of sold cars')
 xyz.xlabel('Dealership')
 xyz.title('Number of cars sold by dealers A,B,C,D,E')
 xyz.xticks(rotation=0)
 xyz.legend()
 xyz.show()
```
